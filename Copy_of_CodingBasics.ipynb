## Generating a dataset of random values

Let's generate a set of random numbers. In python any calculation/operation with numbers starts with **np** followed by another instruction. In our case we want random numbers from a normal (i.e. Gaussian) distribution.
The statement (i.e. instruction) looks like this: np.random.normal(XXXX)

In the parenthesis we tell the statement the specifics. A Gaussian distribution has a mean and a standard deviation (*std*), so we need to specify this. The full statement looks like this **np.random.normal(mean, std, length)**.

We will generate numbers from a normal distribution with a mean of 25, a standard deviation of 3 and a length of 20. Length is how many data points we want.

We need to assign this ramdom numbers to what is called a *variable*. <u> Use **your** initials as the variable name </u>. For me it will be MA. We can later do mathematical operations on this *variable* or plot it.

Finally, to see the random numbers you generated use the **print** statement to see the values in that variable.

```python
MA = np.random.normal(25, 3, 20)
print(MA)
```

Try this code in the cell below. After running the code with my initials (MA) remember to modify it to have **your** initials as the name of the variable.

````python
YL = np.random.normal(25, 3, 20)
print(YL)
````

The next step is to plot the *variable* you just created. We plot using the library **plt**. This is followed by what we would like done, it could be the plot (**plot**), adding labels to the axes (**ylabel**), limiting the values of the axis (**ylim**), etc.

To plot the data as lines use - or -- or -. types
To plot the location of your data points, use markers such as o or s or ^

```python
plt.plot(MA, '-o')
plt.xlabel('Point #')
plt.ylabel('Value')
plt.title('Some random values')
plt.show()
```

````python
plt.plot(YL, '-o')
plt.xlabel('Point #')
plt.ylabel('Value')
plt.title('Some random values')
plt.show()
````

You can see that the components of the variable/vector **MA** are plotted as interconnected dots (that is the meaning of the `-o` format specifier) as a function of their index in the array. Note that indexes in *Python* start at zero.

Now define a **new** variable with <u>20 data points</u>, but this time with different mean and standard deviation. <u>You</u> choose these values.
Plot these data as you did above but in red. To change the colour, in the plot statement add *color='r'* after specifying the line and marker ('-o'), with a comma in between these.   

````python
NEW = np.random.normal(50, 2, 20)
plt.plot(NEW, '-o', color='red')
plt.xlabel('Point #')
plt.ylabel('Value')
plt.title('Some random values')
plt.show()
````

The final step is to do a mathematical operation on these two variables and plot all three.

Make a new variable to be called IWON which is the multiplication of the two random datasets you created. For my example, it looks like this

```python
IWON= MA * NEW
```

**Print this new variable**

````python
IWON= YL * NEW
print(IWON)
````

We will now make a plot that contains three plots of all the variables we created.

For that we will use the statement **subplot**. It will have the following form. plt.subplot(number of rows, number of columns, position of the plot).

For example, if I want to have three plots side by side the statement is:
**plt.subplot(1,3,2)**
The first two numbers define the plot *number of rows=1* and *number of columns*=3.
The third number (2 in this example) is where you want the next statement to plot. There are four positions for plots, a '2' means the 2nd position.

```python
plt.subplot(1,3,1)
plt.plot(MA,'-o')
plt.title('Mean of 25')

plt.subplot(1,3,2)
plt.plot(NEW,'-o',color='r')
plt.title('Mean of 50')

```

**Plot the data of the multiplied variables in the third plot location. Add titles and axis labels to ALL plots**

````python
plt.subplot(1,3,1)
plt.plot(YL,'-o')
plt.title('Mean of 25')

plt.subplot(1,3,2)
plt.plot(NEW,'-o',color='r')
plt.title('Mean of 50')

plt.subplot(1,3,3)
plt.plot(IWON,'-o', color='black')
plt.title('multiplied variables in YL and NEW')
````
